```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import find_peaks

data = pd.read_csv('/home/mar/Documents/TFM/GitHub/HB_PublicData/2-HB-DEGs-on-Liver-Communities/2_FindMolTiCommunities/Find_num_rand/dif_num_rand.csv',sep=';',index_col=0)
```

```python
sns.set_style("whitegrid")
g = sns.relplot(x="num_randomizations", y="num_clusters", kind="line", data=data)
g.fig.suptitle('Visualizing different settings')
g.fig.autofmt_xdate()
plt.savefig('num_randomizations.png', bbox_inches='tight')
```

```python
diff = data.diff()
diff = sns.relplot(x="num_randomizations", y="num_clusters", kind="line", data=diff)
diff.fig.suptitle('Visualizing differences - pandas.Series.diff()')
diff.fig.autofmt_xdate()
plt.savefig('num_rand_diff.png', bbox_inches='tight')
```
