```python
#import metadegs
degs = pd.read_csv('/home/mar/Documents/TFM/GitHub/multiAffinity/additional_out/metaDEGs/metaDEGs/metaDEGs.txt',sep=',')
degs = degs.rename(columns={'Name': 'metaDEGs', 'Score': 'RRA Score'})
degs['metaDEGs'] = degs['metaDEGs'].astype('str')
degs
```

```python
#import affinity correlation
affinity = pd.read_csv('/home/mar/Documents/TFM/GitHub/multiAffinity/additional_out/MultiAffinity/Affinity_Corr.txt',sep=',')
affinity = affinity.rename(columns={'Genes': 'metaDEGs', 'Corr': 'Affinity Corr'})
affinity['metaDEGs'] = affinity['metaDEGs'].astype('str')
affinity
```
